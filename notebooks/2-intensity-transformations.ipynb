## 1. Negative Transformation Function

The negative of an image with intensity levels in the range [0, L-1] is obtained by using the nagative transformation function expressed by  **s = (L-1) - r** , where the values of pixels, before and after processing, is denoted by r and s respectively, and L-1 the higest intesity value.


```python
# In this case, intesity levels are in [0, L-1] = [0, 1].
# Because the image used is a 32-bit floating point grayscale image.
# So in this case l = L-1 = 1
def reverse_intensity(r,l=1):
    s = l - r
    return s
```

```python
def imtransform_negative1(img):
    rows = img.shape[0]
    cols = img.shape[1]
    negative = np.empty((rows,cols,3),dtype=img.dtype)
    for i in range(rows):
        for j in range (cols):
            negative[i,j] = reverse_intensity(img[i,j])
    return negative

def imtransform_negative2(img): # Transforms RGB Image into its negative
    rows = img.shape[0]
    cols = img.shape[1]
    negative = np.empty((rows,cols,3),dtype=img.dtype)
    for i in range(rows):
        for j in range (cols):
            negative[i,j] = reverse_intensity(img[i,j],255)
    return negative
```

```python
def display_figure(img1,img2,title1,title2):
    plt.figure(1,figsize=(10,10))
    plt.subplot(121)
    plt.imshow(img1)
    plt.title(title1)
    plt.xticks([]),plt.yticks([])

    plt.subplot(122)
    plt.imshow(img2)
    plt.title(title2)
    plt.xticks([]),plt.yticks([])
    plt.show()
    
```

```python
display_figure(img,imtransform_negative1(img),'Original','Negative')
display_figure(myimg,imtransform_negative2(myimg),'Original','Negative')
```

## 2. Square Root Transformation Function 

```python
def sqrt(r):
    return r**(1/2)
```

```python
def imtransform_sqrt(img):
    rows = img.shape[0]
    cols = img.shape[1]
    transform = np.empty((rows,cols,3),dtype=img.dtype)
    for i in range(rows):
        for j in range (cols):
            transform[i,j] = sqrt(img[i,j])
    return transform
```

```python
display_figure(img,imtransform_sqrt(img),'Original','Sqrt Transform')
```

## 3. Log Transformations

```python
# s = c log(1+r)
# Here we choose c = 1 
def log(r):
    s = 1*np.log10(1+r)
    return s
```

```python
def imtransform_log(img):
    rows = img.shape[0]
    cols = img.shape[1]
    transform = np.empty((rows,cols,3),dtype=img.dtype)
    for i in range(rows):
        for j in range (cols):
            transform[i,j] = log(img[i,j,:])
    return transform
```

```python
display_figure(img,imtransform_log(img),'Original','Log Transform')
```

## 4. Power-Law or Gamma Transformations

```python
# s = cr**g
# We choose c = 1
def gamma(r,g):
    s = r**g
    return s
```

```python
def imtransform_gamma(img,g=1):
    rows = img.shape[0]
    cols = img.shape[1]
    transform = np.empty((rows,cols,3),dtype=img.dtype)
    for i in range(rows):
        for j in range (cols):
            transform[i,j] = gamma(img[i,j],g)
    return transform
```

```python
display_figure(img,imtransform_gamma(img),'Original','Gamma = 1')
```

```python
display_figure(img,imtransform_gamma(img,0.04),'Original','Gamma = 0.004')
```
