# Getting Started

## Image Processing with Python Matplotlib and Numpy

### Loading Modules

```python
import numpy as np
import matplotlib.pyplot as  plt
import matplotlib.image as mpimg
```

### Loading RGB Image

```python
img = mpimg.imread("data/home.jpg")
img_1 = mpimg.imread("data/butterfly.jpg")
```

### Some Image Properties

```python
img.ndim
```

```python
print("Image Dimension: ",img.ndim)
print("Image Shape: ",img.shape)
```

#### Min and Max Values for R, G and B Channels

```python
print("R Channel:")
print("---------")
print("Min = ",img[:,:,0].min())
print("Max = ",img[:,:,0].max())
```

```python
print("G Channel:")
print("---------")
print("Min = ",img[..., 1].min())
print("Max = ",img[..., 1].max())
```

```python
print("B Channel:")
print("---------")
print("Min = ",img[..., 2].min())
print("Max = ",img[..., 2].max())
```

### Plotting Image Basics

```python
plt.figure(1,figsize=(10,10))

plt.subplot(121)
plt.title("Home")
plt.xticks([]) # hides x-axis ticks
plt.yticks([]) # hides y-axis ticks
plt.imshow(img)

plt.subplot(122)
plt.title("Butterfly")
plt.xticks([]) # hides x-axis ticks
plt.yticks([]) # hides y-axis ticks
plt.imshow(img_1)

plt.show()
```
