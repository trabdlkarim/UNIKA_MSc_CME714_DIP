# **Color Space Conversion**

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg

myimg = mpimg.imread("data/butterfly.jpg")
```

## **1. RGB to Grayscale Conversion**

#### Weighted Method or Luminosity Method

```python
def get_distance(pixel,w=[1/3,1/3,1/3]):
    r,g,b = pixel[0],pixel[1],pixel[2]
    w0,w1,w2=w[0],w[1],w[2]
    return ((r**2)*w0+(g**2)*w1+(b**2)*w2)**0.5
```

```python
rgb = [100,225,53]
get_distance(rgb)
```

#### Conversion Function Using the get_distance Function

```python
def convert_RGB2GRAY(img,weights=[0.6,0.3,0.1]):
    m=img.shape[0]
    n=img.shape[1]
    gray_img = np.zeros((m,n),dtype=np.uint8)
    for i in range(m):
        for j in range(n):
            gray_img[i,j]=get_distance(img[i,j,:],weights)
    return gray_img
```

```python
plt.figure(2,figsize=(10,5))
```

#### Plotting Images 

```python
gray_scale = convert_RGB2GRAY(myimg)

plt.figure(1,figsize=(10,5))
plt.subplot(121)
plt.xticks([]),plt.yticks([])
plt.title("RGB")
plt.imshow(myimg)

plt.subplot(122)
plt.title("Grayscale")
plt.xticks([]),plt.yticks([])
plt.imshow(gray_scale,cmap='gray')
plt.show()
```

## **2. Grayscale to  Binary (Black & White) Image Conversion**

```python
def convert_GRAY2BW(grayscale,threshold=120):
    m=grayscale.shape[0]
    n=grayscale.shape[1]
    img_bw = np.empty((m,n),dtype=grayscale.dtype)
    for i in range(m):
        for j in range(n):
            if grayscale[i,j] > threshold:
                img_bw[i,j]=1
            else:
                img_bw[i,j]=0
    return img_bw
```

#### Plotting Images

```python
plt.figure(2,figsize=(10,5))
plt.subplot(121)
plt.xticks([]),plt.yticks([])
plt.title("Grayscale")
plt.imshow(gray_scale,cmap='gray')

plt.subplot(122)
plt.title("Binary Image")
plt.imshow(convert_GRAY2BW(gray_scale),cmap='gray')
plt.xticks([]),plt.yticks([])

plt.show()
```
